```python
#import data analysis libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
#Read TSV file into pandas dataframe, only use victim and suspect age group and sex columns
df = pd.read_csv("NYPD_Complaint_Data_Historic.tsv", sep="\t", usecols=['SUSP_AGE_GROUP', 
                 'SUSP_SEX', 'VIC_AGE_GROUP', 'VIC_SEX'])
```

```python
#Convert age group and sex to category datatype and filter out any illogical entries
df['SUSP_AGE_GROUP'] = pd.Categorical(df['SUSP_AGE_GROUP'], 
                       categories=['UNKNOWN', '25-44', '18-24', '45-64', '<18', '65+'])
df['SUSP_SEX'] = pd.Categorical(df['SUSP_SEX'])
df['VIC_AGE_GROUP'] = pd.Categorical(df['VIC_AGE_GROUP'], 
                      categories=['UNKNOWN', '25-44', '18-24', '45-64', '<18', '65+'])
df['VIC_SEX'] = pd.Categorical(df['VIC_SEX'])
```

```python
#See how many data points and unique values
df.describe()
```

```python
#Bar plot for each combination of (Victim, Suspect), (Age, Sex)
fig, ax  = plt.subplots(2, 2, sharey='row', figsize=(7, 7))
df.SUSP_AGE_GROUP.value_counts().plot(kind='bar', ax = ax[0, 0])
ax[0, 0].set_title("Count Vs Suspect Age Group"); ax[0, 0].set_xlabel("Age Group"); ax[0, 0].set_ylabel("Count")
df.VIC_AGE_GROUP.value_counts().plot(kind='bar', ax = ax[0, 1])
ax[0, 1].set_title("Count Vs Victim Age Group"); ax[0, 1].set_xlabel("Age Group"); ax[0, 1].set_ylabel("Count")
df.SUSP_SEX.value_counts().plot(kind='bar', ax = ax[1, 0])
ax[1, 0].set_title("Count Vs Suspect Sex"); ax[1, 0].set_xlabel("Sex"); ax[1, 0].set_ylabel("Count")
df.VIC_SEX.value_counts().plot(kind='bar', ax = ax[1, 1])
ax[1, 1].set_title("Count Vs Victim Sex"); ax[1, 1].set_xlabel("Sex"); ax[1, 1].set_ylabel("Count")
fig.tight_layout(h_pad=2)
plt.show()
```
